```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'BMW sales data (2010-2024) (1).csv'
df = pd.read_csv(file_path)
```

```python
# Display basic information about the dataset
print("Dataset Info:")
df.info()
print("\nFirst 5 rows:")
print(df.head())
```

```python
df = df.dropna()
```

```python
# Example: Calculate total sales volume by year
sales_by_year = df.groupby('Year')['Sales_Volume'].sum().reset_index()
```

```python
# Example: Total sales by model
sales_by_model = df.groupby('Model')['Sales_Volume'].sum().reset_index().sort_values('Sales_Volume', ascending=False)
```

```python
# Example: Average price by region
avg_price_by_region = df.groupby('Region')['Price_USD'].mean().reset_index().sort_values('Price_USD', ascending=False)
```

```python
# Print some summaries
print("\nTotal Sales by Year:")
print(sales_by_year)
print("\nTotal Sales by Model:")
print(sales_by_model)
print("\nAverage Price by Region:")
print(avg_price_by_region)
```

```python
# Plot 1: Line plot of Total Sales Volume over Years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Sales_Volume', data=sales_by_year, marker='o')
plt.title('Total BMW Sales Volume by Year (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Total Sales Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 2: Bar plot of Total Sales by Model
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Sales_Volume', data=sales_by_model)
plt.title('Total BMW Sales Volume by Model')
plt.xlabel('Model')
plt.ylabel('Total Sales Volume')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 3: Bar plot of Average Price by Region
plt.figure(figsize=(10, 6))
sns.barplot(x='Region', y='Price_USD', data=avg_price_by_region)
plt.title('Average BMW Price by Region')
plt.xlabel('Region')
plt.ylabel('Average Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot 4: Box plot of Mileage by Fuel Type (example of distribution)
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel_Type', y='Mileage_KM', data=df)
plt.title('Mileage Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Mileage (KM)')
plt.show()
```

```python
# Plot 5: Count plot of Sales Classification by Transmission
plt.figure(figsize=(8, 5))
sns.countplot(x='Sales_Classification', hue='Transmission', data=df)
plt.title('Sales Classification by Transmission')
plt.xlabel('Sales Classification')
plt.ylabel('Count')
plt.show()
```

```python
# Additional: Correlation heatmap (select numeric columns)
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```
